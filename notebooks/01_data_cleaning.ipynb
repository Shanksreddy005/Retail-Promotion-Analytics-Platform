# 01. Data Cleaning & Validation
This notebook documents the cleaning process and data quality validations for the Retail Promotion Analytics Platform.

```python
import pathlib
import pandas as pd
import numpy as np

base_dir = pathlib.Path().resolve().parent
raw_dir = base_dir / 'data' / 'raw'
processed_dir = base_dir / 'data' / 'processed'
```

## Load Datasets

```python
df_orders = pd.read_csv(raw_dir / 'olist_orders_dataset.csv')
df_order_items = pd.read_csv(raw_dir / 'olist_order_items_dataset.csv')
df_customers = pd.read_csv(raw_dir / 'olist_customers_dataset.csv')
print(f'Orders: {df_orders.shape}, Items: {df_order_items.shape}, Customers: {df_customers.shape}')
```

## Missing Value Checks

```python
print('Orders null counts:\n', df_orders.isnull().sum())
print('\nCustomers null counts:\n', df_customers.isnull().sum())
```

## Deduplication

```python
df_orders = df_orders.drop_duplicates()
df_customers = df_customers.drop_duplicates()
print(f'Deduplicated Shapes - Orders: {df_orders.shape}, Customers: {df_customers.shape}')
```
