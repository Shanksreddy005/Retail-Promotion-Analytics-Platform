# 02. Exploratory Data Analysis (EDA)
In this notebook, we explore the promotion dataset to answer key business questions:
1. Which promotions generate the highest revenue?
2. How do promotions affect purchase frequency?
3. How responsive are different regions (states) to promotions?
4. What is the impact of different discount levels on Average Order Value (AOV)?

```python
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

base_dir = pathlib.Path().resolve().parent
processed_dir = base_dir / 'data' / 'processed'
df_orders = pd.read_csv(processed_dir / 'processed_orders.csv')
print(df_orders.info())
```

## 1. Highest Revenue Promotions

```python
revenue_promo = df_orders.groupby('promotion_id')['order_value'].sum().reset_index()
revenue_promo = revenue_promo.sort_values(by='order_value', ascending=False)
print('Total Revenue by Promotion ID:\n', revenue_promo)

sns.barplot(data=revenue_promo, x='promotion_id', y='order_value')
plt.title('Total Revenue by Promotion')
plt.show()
```

## 2. Promotion Effect on Purchase Frequency

```python
freq_promo = df_orders.groupby('promotion_id')['repeat_customer_flag'].mean().reset_index()
print('Repeat Customer Rate by Promotion ID:\n', freq_promo)

sns.barplot(data=freq_promo, x='promotion_id', y='repeat_customer_flag')
plt.title('Repeat Customer Rate by Promotion Group')
plt.show()
```

## 3. Regional Responsiveness

```python
regional = df_orders.groupby(['customer_state', 'promotion_flag'])['order_value'].mean().unstack()
regional['uplift_pct'] = ((regional[1] - regional[0]) / regional[0]) * 100
regional = regional.sort_values(by='uplift_pct', ascending=False)
print('Regional AOV Promotion Uplift %:\n', regional.head(10))

regional['uplift_pct'].head(15).plot(kind='bar')
plt.ylabel('AOV Uplift (%)')
plt.title('Top States by Promotional AOV Uplift %')
plt.show()
```

## 4. Discount Impact on AOV

```python
aov_discount = df_orders.groupby('promotion_id')['order_value'].mean().reset_index()
print('Average Order Value (AOV) by Promotion ID:\n', aov_discount)

sns.boxplot(data=df_orders, x='promotion_id', y='order_value')
plt.ylim(0, 300) # limit y-axis for better visibility
plt.title('Order Value Distribution by Group')
plt.show()
```
